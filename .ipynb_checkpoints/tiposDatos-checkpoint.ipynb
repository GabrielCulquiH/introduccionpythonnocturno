## 2.1 Literales

* Pytohon tiene una serie de tipos de datos integrados en el propio lenguaje
* Los literales son expresiones que generan objetos de estos tipos
* Estos objetos, según sus tipos pueden ser:
    * Simples o compuestos
    * Mutables e inmutables

### Literales simples
- Enteros
- Decimales o puntos flotantes
- Booleanos


```python
print(4) # Número entero
print(4.3) # Número en punto flotante
print('Hello world') # Objeto de tipo string
print(False) # Objeto de tipo booleano, puede ser solo falso o verdadero (True/False)
```

### Literales compuestos
* Tuplas
* Listas
* Diccionarios
* Conjuntos

```python
print ([1,2,3,3,7]) #Lista - mutable
print ({'Nombre' : 'John Doe', "edad": 30, "soltero":True}) #diccionario - mutable
print ({1,2,3,3}) #conjunto de datos - mutable (sin datos repetidos)
print ((4,5)) #tupla - inmutable (no se puede cambiar para nada)
2,4 #Tupla - inmutable (otra forma de crear)
```

## 2.2 Variables
* Referencias a objetos
* Las variables y los objetos se almacenan en espacios de memoria distinta
* Las variables siempre referencian a objetos y nunca a otras variables
* Los objetos sí pueden referenciar a otros objetos. Ejemplos: listas
* Sentencia de asignación:

<"nombre_variable"> '=' <"objeto">


```python
# Asignación de variables
a=5
print (a)
```

```python
a = 1 # Entero
b = 4.0 # Flotante
c = "ITQ" # String
d = 10+1j # Complejo
e = False # Booleano
f = None # None

#Imprimir variables y ver su tipo
print(a)
print(type(a))

print(b)
print(type(b))

print(c)
print(type(c))

print(d)
print(type(d))

print(e)
print(type(e))

print(f)
print(type(f))
```

* Las variables no tienen tipo
* Las variables apuntan a objetos que sí lo tienen
* Dado que Python es un lenguaje de tipado dinámico la misma variable puede apuntar, em momentos diferentes de la ejecución del programa, a objetos de diferente tipo

```python
a = 3
print (a)
print(type(a))

a=100
print (a)
print(type(a))

a="Pablo"
print (a)
print(type(a))

a=4.5
print (a)
print(type(a))
```

* Esto se llama garbage colection, es justamente la eliminación progresiva de los espacios de memoria que un objeto ocupa.

```python
a=3
print (id(a))

a= "Pablo García"
print(id(a))

a=4.5
print(id(a))
```

### Nombres de las variables
* No se aceptan números delante del nombre de las variables
* Por convención usar el SnakeCase y no el CamelCase, es decir usar "_" para los espacios entre palabras en el nombre de la variable
* Tener cuidado, el lenguaje identifica mayúsculas y minúsculas
* Nombres deben ser descriptivos
* Hay palabras o métodos reservados -> se denominan Built-ins y KeyWords
  * NO ASIGNAR NOMBRES RESERVADOS

```python
print(pow(3,2)) #Potencia
```

```python
print(pow(3,2))

pow=1 #built-in con valor reasignado
print(pow(3,2))

#El lenguaje no comprende que hacer con la reasignación
```

```python
#Redefine pow como una nueva función
def pow(a,b):
    return a+b

print(pow(3,2)) 
```

## 2.3 Tipos de datos básicos

### Bool

* Tiene 2 posibles valores: 'True' o 'False'

```python
a=False
b=True

print(a)
print(type(a))

print(b)
print(type(b))
```

* "True" y "False" también son objetos que se guardan en caché, al igual que los enteros pequeños

```python
a=True
b=False

print(id(a))
print(id(b))
print(a is b)
print(a == b)
```

### Números

```python
print(2) #Enteros, sin parte fraccional
print(3.4) #Flotantes, con parte fraccional
print(2+4j) #Complejos
print(1/2) #Racionales
```

* Diferentes representaciones: base 10,2,8,16

```python
x = 58 #decimal
z = 0b00111010 #binario
w = 0o72 #octal
y = 0x3A #hexadecimal
print(x == y == z == w)

#Todos representan el mismo número
```

### Strings

* Cadenas de caracteres
* Son secuencias: la posición de los caracteres es importantes.
* Son inmutables: las operaciones sobre strings no cambian el string original.

```python
s = 'John "ee" Doe'

print (s[0]) # Primer caracter del string
print (s[-1]) # Último caracter del string
print (s[1:8:2]) #Substring desde el segundo caracter (inclusive) (último valor excluyente)
print (s[::2]) #Todo el string
print (s + "e") # Concatenación
```

## 2.4 Conversión entre tipos

* A  veces queremos que un objeto sea de un tipo específico
* Podemos obtenerobjetos de un tipo a partir de objetos de un tipo diferente (*casting*).

```python
a = int(2.8) # a será 2
print(a)
print(type(a))

b = int(3) # b será 3
print(b)
print(type(b))

c = float(1) # c será 1.0
print(c)
print(type(c))

d = float("3") # d será 3.0
print(d)
print(type(d))

e = str(2) # e será 2
print(e)
print(type(e))

f = str(3.0) # f será '3.0'
print(f)
print(type(f))

g = bool("a") # g será True
print(g)
print(type(g))

h = bool("") # h será False
print(h)
print(type(h))

i = bool(3) # i será True
print(i)
print(type(i))

j = bool(0) # j será False
print(j)
print(type(j))

k = bool(None) # k será False
print(k)
print(type(k))
```

```python
print (7/4) #División convencional, resultado tipo float
print (7//4) # División entera. Resultado tipo "int"
print (int(7/4)) # División convencional. Conversión de float a int
```

## 2.5 Operadores 
* Combinación de valores, variables y operadores
* Operadores y operandos

#### Operadores aritméticos
| Operador | Desc |
|:---------|:-----|
| a + b    | Suma |
| a - b    | Resta |
| a / b    | División |
| a // b   | División Entera |
| a % b    | Modulo / Resto |
| a * b    | Multiplicacion |
| a ** b   | Exponenciación |

```python
x = 3
y = 2

print('x+y = ',x+y)
print('x-y = ',x-y)
print('x*y = ',x*y)
print('x/y = ',x/y)
print('x//y = ',x//y)
print('x%y = ',x%y)
print('x**y = ',x**y)
```

#### Operadores de comparación
| Operador | Desc |
|:---------|:-----|
| a > b    | Mayor |
| a < b    | Menor |
| a == b    | Igual |
| a != b   | Diferente |
| a >= b    | Mayor o igual |
| a <= b    | Menor o igual |


```python
x = 26
y = 24
print('x>y es ', x>y)
print('x<y es ', x<y)
print('x==y es ', x==y)
print('x!=y es ', x!=y)
print('x>=y es ', x>=y)
print('x<=y es ', x<=y)

x = 10
y = 10
print('x>y es ', x>y)
print('x<y es ', x<y)
print('x==y es ', x==y)
print('x!=y es ', x!=y)
print('x>=y es ', x>=y)
print('x<=y es ', x<=y)
```

#### Operadores Lógicos

| Operador | Desc |
|:---------|:-----|
| a and b    | True, si ambos son True |
| a or b    | True, si al menos uno es True |
| a ^ b    | XOR - True, si solo uno de los dos es True |
| a not b   | Negación |


Más sobre tablas de verdad => [enlace](https://en.wikipedia.org/wiki/Truth_table)

```python
x=True
y=False
print('x and y es :', x and y)
print('x or y es :', x or y)
print('x xor y es :', x ^ y)
print('not x es :', not x)
```

#### Operadores Bitwise / Binarios
| Operador | Desc |
|:---------|:-----|
| a & b    | AND binario |
| a | b    | OR binario |
| a ^ b    | XOR binario |
| -a  | NOT binario |
| a >> b  | Desplazamiento binario derecha |
| a << b  | Desplazamiento binario izquierda |

```python
x = 0b01100110
y = 0b00110011
print("Not x = " + bin(~x))
print("x and y = " + bin(x & y))
print("x or y = " + bin(x | y))
print("x xor y = " + bin(x ^ y))
print("x << 2 = " + bin(x << 2))
print("x >> 2 = " + bin(x >> 2))
```

#### Operadores de asignación
| Operador | Desc |
|:---------|:-----|
| =    | Asignación |
| +=    | Suma y asignación |
| -=   | Resta y asignación |
| *=  | Multiplicación y asignación |
| /=  | División y asignación |
| %=    | Módulo y asignación|
| //=  | División entera y asignación |
| **=  | Exponencial y asignación |
| &=  | AND y asignación |
| =  | OR y asignación |
| ^=  | XOR y asignación |
| >>=  | Desplazamiento derecha y asignación |
| <<=  | Desplazamiento izquierda y asignación |

```python
a = 8
a *= 5 # a = a * 5
a += 2 # No existe a++, ni ++a,a--,--a
print(a)
b = 10
b -= 5 # b = b - 
print(b)
```

#### Operadores de Identidad
| Operador | Desc |
|:---------|:-----|
| a is b    | True si los dos referencian al mismo objeto |
| a is not b   | True si ambos no referencian al mismo objeto |


```python
a=123
b=a
print(a is b)
print(a is not b)

a=500
b=500
print(a is b)
print(a is not b)
```

#### Operadores de Pertenencia
| Operador | Desc |
|:---------|:-----|
| a in b    | True si a se encuentra en b |
| a not in b   | True si a no se encuentra en b |


```python
x = "Hello world"
y = {1:'a',2:'b'}
print('l' in x) #True
print("Hello" not in x) #True
print(1 in y) # True
print('a' in y) # False
```

Elaborar un programa que permita calcular el volumen de un cilindro solicitando al usuario que ingrese los valores por teclado, radio y altura del cilindro. El resultado debe mostrarse en pantalla, no se debe utilizar condicionales ni funciones matemáticas, solo operaciones aritméticas básicas y lectura de datos por teclado. Fórmula del volumen: $$V = pi * radio^2 * altura $$

r = radio

```python
#Solución
rad = int(input("Ingrese radio en cm: "))
print(f"El valor ingresado es {rad}")
alt = int(input("Ingrese altura en cm: "))
print(f"El valor ingresado es {alt}")
print("El valor de PI es 3.1416")
print(f"El volumen del cilindro en cm3 es: {3.1416*(rad**2)*alt}")
```
