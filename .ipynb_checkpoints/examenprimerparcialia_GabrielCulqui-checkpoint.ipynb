1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$

### SOLUCIÓN:


```python
print("Calculadora de IMC")
print("Para el cálculo se requiere conocer su estatura en metros y su peso en kg")

estat = float(input("Ingrese su estatura (m): "))
peso = float(input("Ingrese su peso (kg): "))

print(f"Su IMC es: {peso/(estat**2)}")
```

2. Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y2,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz

### SOLUCIÓN (1):


```python
print("Calculadora de distancia entre puntos")
print("Para el cálculo se requiere conocer las coordenadas (x, y) de ambos puntos")

x1 = float(input("Ingrese la coordenada x del punto 1: "))
y1 = float(input("Ingrese la coordenada y del punto 1: "))
x2 = float(input("Ingrese la coordenada x del punto 2: "))
y2 = float(input("Ingrese la coordenada y del punto 2: "))

print(f"La distancia entre los puntos es: {(((x2-x1)**2)+((y2-y1)**2))**0.5}")
```

### SOLUCIÓN (2):

```python
import math
print("Calculadora de distancia entre puntos")
print("Para el cálculo se requiere conocer las coordenadas (x, y) de ambos puntos")

x1 = float(input("Ingrese la coordenada x del punto 1: "))
y1 = float(input("Ingrese la coordenada y del punto 1: "))
x2 = float(input("Ingrese la coordenada x del punto 2: "))
y2 = float(input("Ingrese la coordenada y del punto 2: "))

var = (((x2-x1)**2)+((y2-y1)**2))

print(f"La distancia entre los puntos es: {math.sqrt(var)}")
```
