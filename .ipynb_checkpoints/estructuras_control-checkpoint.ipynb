 ## Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’ al finalizar cada sentencia

```python
 # # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x: "))
#Implemento una condicional para saber si x es mayor a 0

if x > 0:
    print('número positivo')
else:
    print('número no positivo')
    
```

```python
x = 26
y = [10, 20, 30]
print(x > 0)
1
print(x > 0 and x < 10) #Cuando las dos condiciones es verdadero
print(x is not y) #Compara identificadores, compara espacios de memoria no valores
print(x in y)
```

**Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave *elif***

```python
x = 2026 #Ingrese dato
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo') #Se vuelve a realizar la sentencia, elif es 'si no si'
elif x < 0:
    print('Valor negativo')
else:
    print('Error, revisar datos')
```

## SWITCH

```python
import numpy as np

a = 8
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
    case _:
        print("ERROR DE NUMERO")


```

```python
a = 'x'
match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

## Iteración (bucles)

* Repetición de un bloque de código.
* La terminación del bucle depende del tipo de bucle.
* Dos tipos: while y for.
  
**Bucle ‘while’**
* Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta
que una condición evalue a False).
* Si la condición evalua a False desde el principio, el bloque de código nunca se ejecuta.
* Cuidado con los bucles infinitos.

Formato general:

while test: # Mientras se cumple la condición

    statements # Instrucciones a ejecutar

```python
# Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
 # Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador = contador + 1
    indice = indice + 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

## Bucle ‘for’

* Permite recorrer los items de una sequencia o un objeto iterable.
* Funciona en strings, listas, tuplas, etc.

Formato general:
for item in objeto:       # Asigna los items del objeto a la variable item en cada iteración
statements                # Instrucciones a ejecutar


for item range(inicio,fin,2):

    statements

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

# for (pelicula=0; pelicula<leng(peliculas);pelicula++)
for pelicula in peliculas:
    print(pelicula)
```

* Se usa para iterar un número preestabecido de veces (counted loops):

```python
for i in range(10):
    print(i)
```

* range es útil en combinación con len porque permite accceder a los lementos de una secuencia por posición

```python
nombre = "Pablo"

for i in range(len(nombre)):
    print(nombre[i])
```

```python
for letra in nombre:
    print(letra)
```

* Iteración de tuplas

```python
tuplas = [(1,2), (3,4), (5,6)]
for a,b in tuplas:
    print(a,b)
```

* Iteración de diccionarios. Se iteran las claves

```python
diccionario = {"a":1,"b":2,"c":3}

for key in diccionario:
    print(f"{key} => {diccionario[key]}")
```

*  Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items
y values

```python
for key, value in diccionario.items():
    print(f"{key} => {value}")
```

```python
for value in diccionario.values():
    print(value)
```

* La variable item en la cabecera del for puede ser cualquier expresión que sea válida como
parte izquierda de una asignación convencional

```python
 for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
```

## EJERCICIO

Crear un programa en Python donde el usuario ingrese un número por teclado, mediante el ciclo for verificar si los números sucesivos previos son pares e impares.

```python
#Solución

num = int(input("Ingrese número: "))

lista = range(num+1)

for i in lista:
    match i:
        case _ if i==0:
            print(i, "Es cero")
        case _ if i%2 == 0:
            print(i, "Es par")
        case _ if i%2 != 0:
            print(i, "Es impar")
    
    
    
```

## EJERCICIO

Desarrollar un programa en Python donde el usuario ingrese un número entero positivo, verificar si el número ingresado es primo o no.

```python
#Solución
num= int(input("Ingrese el número a verificar: "))
mod = []
for i in range(1,num+1):
    mod.append(num%i)

lista = mod[1:-1]

if 0 in lista:
    print("El número ingresado no es primo")

else: 
    print("El número es primo")
    
print(lista)
```
