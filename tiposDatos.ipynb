## 2.1 Literales

* Pytohon tiene una serie de tipos de datos integrados en el propio lenguaje
* Los literales son expresiones que generan objetos de estos tipos
* Estos objetos, según sus tipos pueden ser:
    * Simples o compuestos
    * Mutables e inmutables

### Literales simples
- Enteros
- Decimales o puntos flotantes
- Booleanos


```python
print(4) # Número entero
print(4.3) # Número en punto flotante
print('Hello world') # Objeto de tipo string
print(False) # Objeto de tipo booleano, puede ser solo falso o verdadero (True/False)
```

### Literales compuestos
* Tuplas
* Listas
* Diccionarios
* Conjuntos

```python
print ([1,2,3,3,7]) #Lista - mutable
print ({'Nombre' : 'John Doe', "edad": 30, "soltero":True}) #diccionario - mutable
print ({1,2,3,3}) #conjunto de datos - mutable (sin datos repetidos)
print ((4,5)) #tupla - inmutable (no se puede cambiar para nada)
2,4 #Tupla - inmutable (otra forma de crear)
```

## 2.2 Variables
* Referencias a objetos
* Las variables y los objetos se almacenan en espacios de memoria distinta
* Las variables siempre referencian a objetos y nunca a otras variables
* Los objetos sí pueden referenciar a otros objetos. Ejemplos: listas
* Sentencia de asignación:

<"nombre_variable"> '=' <"objeto">


```python
# Asignación de variables
a=5
print (a)
```

```python
a = 1 # Entero
b = 4.0 # Flotante
c = "ITQ" # String
d = 10+1j # Complejo
e = False # Booleano
f = None # None

#Imprimir variables y ver su tipo
print(a)
print(type(a))

print(b)
print(type(b))

print(c)
print(type(c))

print(d)
print(type(d))

print(e)
print(type(e))

print(f)
print(type(f))
```

* Las variables no tienen tipo
* Las variables apuntan a objetos que sí lo tienen
* Dado que Python es un lenguaje de tipado dinámico la misma variable puede apuntar, em momentos diferentes de la ejecución del programa, a objetos de diferente tipo

```python
a = 3
print (a)
print(type(a))

a=100
print (a)
print(type(a))

a="Pablo"
print (a)
print(type(a))

a=4.5
print (a)
print(type(a))
```

* Esto se llama garbage colection, es justamente la eliminación progresiva de los espacios de memoria que un objeto ocupa. 

## Identificadores
* Podemos obtener un identificador único para los objetos referenciados por variables.
* Este identificador se obtiene a partir de la dirección de memoria.


```python
a=3
print (id(a))

a= "Pablo García"
print(id(a))

a=4.5
print(id(a))
```

* Referencias compartidas: un mismo objeto puede ser referenciado por más de una variable

 - Variables que referencian al mismo objeto tienen mismo identificador.

```python
a=4567
print(id(a))
```

```python
b=a
print(id(b))
```

```python
c=4567
print(id(c)) #se guarda en un espacio diferente es otro objeto
```

```python
a=25
b=25

print(id(a))
print(id(b))
print(id(25))
```

```python
#Los números grandes [-5,256] fuera del rango 0-255 
#van a abrir un nuevo ciclo de cuenta internamente, nunca tendrán un mismo identificador
a=258
b=258

print(id(a))
print(id(b))
print(id(258))
```

* Referencia al mismo objeo a través de asignar una variable a otra

```python
a=400
b=a
print(id(a))
print(id(b))
```

### Nombres de las variables
* No se aceptan números delante del nombre de las variables
* Por convención usar el SnakeCase y no el CamelCase, es decir usar "_" para los espacios entre palabras en el nombre de la variable
* Tener cuidado, el lenguaje identifica mayúsculas y minúsculas
* Nombres deben ser descriptivos
* Hay palabras o métodos reservados -> se denominan Built-ins y KeyWords
  * NO ASIGNAR NOMBRES RESERVADOS

```python
print(pow(3,2)) #Potencia
```

```python
print(pow(3,2))

pow=1 #built-in con valor reasignado
print(pow(3,2))

#El lenguaje no comprende que hacer con la reasignación
```

```python
#Redefine pow como una nueva función
def pow(a,b):
    return a+b

print(pow(3,2)) 
```

## 2.3 Tipos de datos básicos

### Bool

* Tiene 2 posibles valores: 'True' o 'False'

```python
a=False
b=True

print(a)
print(type(a))

print(b)
print(type(b))
```

* "True" y "False" también son objetos que se guardan en caché, al igual que los enteros pequeños

```python
a=True
b=False

print(id(a))
print(id(b))
print(a is b)
print(a == b)
```

### Números

```python
print(2) #Enteros, sin parte fraccional
print(3.4) #Flotantes, con parte fraccional
print(2+4j) #Complejos
print(1/2) #Racionales
```

* Diferentes representaciones: base 10,2,8,16

```python
x = 58 #decimal
z = 0b00111010 #binario
w = 0o72 #octal
y = 0x3A #hexadecimal
print(x == y == z == w)

#Todos representan el mismo número
```
