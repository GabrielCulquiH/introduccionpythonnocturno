### Tipos de comentarios
**Comentarios de una línea**

* Texto precedido por el signo #
* Luego le decimos se suele usar para documentar expresiones sencillas

```python
# Esto es una instrucción print
print("Hola mundo") #Esto es una instrucción
```

**Comentarios de varias líneas**
* Texto encapsulado en triple comillas (que pueden ser simples como dobles)
* Se suele usar para documentar bloques de código más significativos

```python
# Creo una función de multiplicación
def producto(x,y): 
    '''
    Esta función recibe dos números x e y como parámetros y devuelve como resultado el producto de los mismos
    '''
    # Creo una variable donde guardo el producto de los números recibidos
    c=x*y
    return c
```

```python
# Llamo a la función
producto(3,7)
```

# Literales, variables y tipos de datos básicos

De forma muy genérica al ejecutarse un program en Python simplemente se realizan dos cosas se realizan operaciones sobre objetos

Estos dos términos son fundamentales

* *Objetos*: cualquier tipo de datos que sean (numéricos, caracteres o datos más complejos)
* *Operaciones*: como manipulamos esos datos. +,-,*,/

Ejemplo:

```python
4*3
```

## Literales

* Pytohon tiene una serie de tipos de datos integrados en el propio lenguaje
* Los literales son expresiones que generan objetos de estos tipos
* Estos objetos, según sus tipos pueden ser:
    * Simples o compuestos
    * Mutables e inmutables

### Literales simples
- Enteros
- Decimales o puntos flotantes
- Booleanos


```python
print(4) # Número entero
print(4.3) # Número en punto flotante
print('Hello world') # Objeto de tipo string
print(False) # Objeto de tipo booleano, puede ser solo falso o verdadero (True/False)
```
