## 2.2 Variables
* Referencias a objetos
* Las variables y los objetos se almacenan en espacios de memoria distinta
* Las variables siempre referencian a objetos y nunca a otras variables
* Los objetos sí pueden referenciar a otros objetos. Ejemplos: listas
* Sentencia de asignación:

<"nombre_variable"> '=' <"objeto">


```python
# Asignación de variables
a=5
print (a)
```

```python
a = 1 # Entero
b = 4.0 # Flotante
c = "ITQ" # String
d = 10+1j # Complejo
e = False # Booleano
f = None # None

#Imprimir variables y ver su tipo
print(a)
print(type(a))

print(b)
print(type(b))

print(c)
print(type(c))

print(d)
print(type(d))

print(e)
print(type(e))

print(f)
print(type(f))
```

```python
a = 3
print (a)
print(type(a))

a=100
print (a)
print(type(a))

a="Pablo"
print (a)
print(type(a))

a=4.5
print (a)
print(type(a))
```

## 2.4 Conversión entre tipos

* A  veces queremos que un objeto sea de un tipo específico
* Podemos obtenerobjetos de un tipo a partir de objetos de un tipo diferente (*casting*).

```python
a = int(2.8) # a será 2
print(a)
print(type(a))

b = int(3) # b será 3
print(b)
print(type(b))

c = float(1) # c será 1.0
print(c)
print(type(c))

d = float("3") # d será 3.0
print(d)
print(type(d))

e = str(2) # e será 2
print(e)
print(type(e))

f = str(3.0) # f será '3.0'
print(f)
print(type(f))

g = bool("a") # g será True
print(g)
print(type(g))

h = bool("") # h será False
print(h)
print(type(h))

i = bool(3) # i será True
print(i)
print(type(i))

j = bool(0) # j será False
print(j)
print(type(j))

k = bool(None) # k será False
print(k)
print(type(k))
```

```python
print (7/4) #División convencional, resultado tipo float
print (7//4) # División entera. Resultado tipo "int"
print (int(7/4)) # División convencional. Conversión de float a int
```

 ## 3.1 Listas
* Una colección de objetos.
* Mutables.
* Tipos arbitrarios heterogeneos.
* Puede contener duplicados.
* No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
* Los elementos van ordenados por posición.
* Se acceden usando la sintaxis: [index].
* Los índices van de 0 a n-1, donde n es el número de elementos de la lista.
* Son un tipo de Secuencia, al igual que los strings; por lo tanto, el orden (es decir, la posición de los objetos de la lista) es importante.
* Soportan anidamiento.
* Son una implementación del tipo abstracto de datos: Array Dinámico.

```python
letras = ['a', 'b', 'c', 'd']
palabras = 'Hola mundo como estas'.split()
numeros = list(range(7))
print(letras)
print(palabras)
print(numeros)
print(type(numeros))
```

### Alias/referencias de listas

* Son mutables y las asignaciones a un objeto alteran el primero
* Pasar listas a funciones puede ser riesgoso
* Clonar o copiar listas

```python
lista = [5,10,15,20]
ref = lista
print(id(lista))
print(id(ref))

ref[2] =64

print((ref))
print((lista))
```

```python
lista  = [2048, 1024, 512, 256]
copia = lista[:]
copia = lista.copy()

copia[2] = 64

print(lista)
print(copia)

print(id(lista))
print(id(copia))
```

```python
#Listas anidadas, se copian por referencia

lista = [2, 4, 16, 32, [34, 10, [5,5]]]
copia = lista.copy()
copia[3] = 20
copia[4][0] = 28
print(copia)
print(lista)
```

```python
lista = [0, 1, [10, 20]] #Crea lista
print(id(lista)) #Id de la lista
lista2 = [10, 20] #Crea nueva lista
lista = [0, 1, lista2] #Anida la lista
print(lista)
copia = lista[:] #.copy() # copia la lista
print(copia) #Imprime la lista
print(id(copia)) #Imprime ID de la lista
print(id(lista[2])) #Imprime el ID del objeto en posición 2 de la lista
print(id(copia[2])) #Imprime el ID del objeto en posición 2 de la copia
copia[2][0] = 40 #Reemplaza valores
print(copia)
print(lista)
```

```python
# # evitar que listas anidadas se copien por referencia
import copy
lista = [2, 4, 16, 32, [34, 10, [5,5]]] # Crea lista
copia = copy.deepcopy(lista) #Copia lista con funcion deepcopy
copia[0] = 454 #Reemplaza valores
copia[4][2][0] = 64 #Reemplaza valores
print(lista)
print(copia)
print(f"{id(lista)} - {id(copia)}") # Imprime id en una misma línea
print(f"{id(lista[4])} - {id(copia[4])}") # Imprime id del objeto en una misma línea
```

## Diccionarios

* Colección de parejas (clave - valor)
*  Son mutables.
* Claves:
    * Cualquier objeto inmutable.
    * Sólo pueden aparecer una vez en el diccionario.
* Valores:
    * Sin restricciones. Cualquier objeto (enteros, strings, listas, etc.) puede hacer de valor.
*  Desde la versión 3.7 están ordenados.
*  Se pueden ver como listas indexadas por cualquier objeto inmutable, no necesariamente por
números enteros.
*  A diferencia de las listas, no son secuencias. Son mappings.


#### Operaciones con diccionarios

* Creación de diccionarios

```python
# Creación simple, usando una expresión literal.
persona = {'CI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}
print(persona)
```

```python
# Unir colecciones
nombres = ['Pablo', 'Manolo', 'Pepe', 'Juan']
edades = [52, 14, 65]
datos = dict(zip(nombres,edades))
print(datos)
# No importa cuantos datos tengan los diccionarios, toma los datos en orden
```

```python
# Creación pasando claves y valores a la función 'dict'
persona2 = dict(nombre='Rosa', apellido='Garcia')
print(persona2)
```

```python
# Creación usando una lista de tuplas de dos elementos.
persona2 = dict([('nombre', 'Rosa'), ('apellido','Garcia')])
print(persona2)
```

```python
 # Creación incremental por medio de asignación (como las claves no existen, se crean nuevos items)
persona = {}
# print(persona['CI'])
persona['CI'] = '11111111D'
persona['Nombre'] = 'Carlos'
persona['Edad'] = 34
persona['CI'] = '222222222'
print(persona)
```

*  Acceso a un valor a través de la clave

```python
persona = {'CI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34, 'Peso':34}
print(persona['Nombre'])
```

```python
 # Acceso a claves inexistentes o por índice produce error
# persona[1]
print(persona['Nombre'])

if 'Trabajo' in persona:

    print(persona['Trabajo'])

else:
    print("Desempleado")
    
persona.get('Trabajo', "En el paro") #obtiene valores
```

* Modificación de un valor a través de la clave

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}
persona['Nombre'] = 'Fernando'
persona['Edad'] += 1
print(persona)
```

* Añadir un valor a través de la clave.


```python
persona = {'CI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}
persona['Ciudad'] = 'Quito'
print(persona)
```

* Eliminación de un valor a través de la clave

```python
del persona['Ciudad'] # del elimina la clave
value = persona.pop('Edad') #pop elimina la clave en el valor final

print(persona)
print(value)
```

* Comprobación de existencia de clave.

```python
print('Nombre' in persona)
print('Apellido' in persona)
```

* Recuperación del valor de una clave, indicando valor por defecto en caso de ausencia.

```python
persona = {'Nombre' : 'Carlos'}
value = persona.get('Nombre')
print(value)
# persona['Estatura']
value = persona.get('Estatura', 180)
print(value)
```

* Anidamiento

```python
persona = {
'Trabajos' : ['desarrollador', 'gestor'],
'Direccion' : {'Calle' : 'Pintor Guayasamin', 'Ciudad' : 'Quito'}
}
print(persona['Direccion'])
print(persona['Direccion']['Ciudad'])
```

* Métodos items, keys y values.

```python
persona = {'CI' : '11111111D', 'Nombre' : 'Carlos', 'Edad' : 34}

print(list(persona.items())) #items crea lista de tuplas clave-dato
print(list(persona.keys())) #keys crea lista de claves
print(list(persona.values())) #values crea lista con los valores de los datos
```

```python
dict_simple = {'ID' : 'XCSDe1194', 'Nombre' : 'Carlos', 'Edad' : 34}
# iterar sobre las claves
print('Claves\n')
for key in dict_simple.keys(): # o dict_simple (recorre todo el diccionario extrayendo las claves, una a la vez)
    print(key)

print('\nValores\n')
# iterar sobre los valores
for value in dict_simple.values():
    print(value)
```

```python
dict_simple = {'ID' : 'XCSDe1194', 'Nombre' : 'Carlos', 'Edad' : 34}
# iterar sobre ambos, directamente con items().
for key, value in dict_simple.items():
    print('Clave: ' + str(key) + ', valor: ' + str(value))
    
for item in dict_simple.items(): # devuelve tuplas
    print('Item: ' + str(item))
    
# usar zip para iterar sobre Clave-Valor
for key, value in zip(dict_simple.keys(),dict_simple.values()):
    print('Clave: ' + str(key) + ', valor: ' + str(value))
```

### Sets (Conjuntos)
Al igual que las listas:
* Colección de elementos.
* Tipos arbitrarios.
* Mutables.
* No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.

A diferencia de las listas:
* No puede tener duplicados.
* Se definen por medio de llaves.
* Los elementos no van ordenados por posición. No hay orden establecido.
* Solo pueden contener objetos inmutables.
* No soportan anidamiento.

Operaciones con conjuntos
* Creación de conjuntos.


```python
set1 = {0, 1, 1, 2, 3, 4, 4}
print(set1)

set2 = {'user1', 12, 2}
print(set2)

set3 = set(range(7))
print(set3)

set4 = set([0, 1, 2, 3, 4, 0, 1])
print(set4)
```

```python
 #Observa la diferencia entre listas y conjuntos

s = 'aabbc'
print(list(s))
print(set(s)) # Elimina los duplicados
```

```python
#Acceder por índice
set1 = {0, 1, 2}
print(set1[0])
# No se puede no hay indice en el conjunto de datos
```

**Unión, intersección y diferencia.**

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4, 42])
# union
print(set1 | set2)

# intersección
print(set1 & set2)

# diferencia
print(set1 - set2) #1-2
print(set2 - set1) #2-1
```

```python
a = [1,2]
b = [2,3]
print(set(a) & set(b)) #intersección de conjuntos
```

```python
#Además de los operadores, que operan únicamente con Sets, también se pueden usar métodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 1, 2}
lista = [1, 3, 3]

print(conjunto.union(lista)) #Unión de conjuntos
print(conjunto.intersection(lista)) #Intersección de conjuntos
print(conjunto.difference(lista)) #Diferencia de conjuntos
```

**Comparación de conjuntos.**

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4])

print(set2.issubset(set1))
print(set1.issuperset(set2)) #Conjunto 1 es padre del 2
print(set1.isdisjoint(set2)) #True si los conjuntos no tienen elementos en común
```

**Pertenencia**

```python
words = {'calm', 'balm'}
print('calm' in words)
```

**Anidamiento**

```python
# Los conjuntos no soportan anidamiento, pero como permite elementos inmutables, se pueden "anidar" tuplas.
nested_set = {1, (1, 1, 1), 2, 3, (1,1,1)}
print(nested_set)
```

**Modificación de conjuntos**

```python
# A través de operador de asignación
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1 |= set2 # set1 = set1 | set2
# set1 &= set2
set1 -= set2
print(set1)
```

```python
# A través de método 'update'.
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1.update(set2)
# set1.intersection_update(set2)
set1.difference_update(set2)
print(set1)
```

```python
# A través de métodos 'add' y 'remove'.
set1 = {'a', 'b', 'c'}
set1.add('d') # Adiciona
set1.remove('a') #Remueve
print(set1)
```
