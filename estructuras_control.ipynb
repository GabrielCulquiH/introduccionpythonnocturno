# Estructuras de control

 ## Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’ al finalizar cada sentencia

```python
 # # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x: "))
#Implemento una condicional para saber si x es mayor a 0

if x > 0:
    print('número positivo')
else:
    print('número no positivo')
    
```

```python
x = 26
y = [10, 20, 30]
print(x > 0)
1
print(x > 0 and x < 10) #Cuando las dos condiciones es verdadero
print(x is not y) #Compara identificadores, compara espacios de memoria no valores
print(x in y)
```

**Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave *elif***

```python
x = 2026 #Ingrese dato
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo') #Se vuelve a realizar la sentencia, elif es 'si no si'
elif x < 0:
    print('Valor negativo')
else:
    print('Error, revisar datos')
```

### Ejercicio 2

Escribir un programa en Python donde el usuario ingrese dos números por teclado y muestre en pantalla la división de los dos números si el divisor es 0 el programa debe mostrar un mensaje, "no se puede dividir un valor para 0".

```python
# Solución
print("Calculadora de división")
n1 = int(input("Ingrese primer número (dividendo): "))
n2 = int(input("Ingrese segundo número (divisor): "))

if n2 == 0:
    print("No se puede dividir un valor para 0")
else:
    print(f"El resultado de la división es: {n1/n2}")
    
```

## SWITCH

```python
import numpy as np

a = 8
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
    case _:
        print("ERROR DE NUMERO")


```
